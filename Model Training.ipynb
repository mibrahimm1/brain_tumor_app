```python
from ultralytics import YOLO
from ultralytics import SAM

model = YOLO("/content/runs/detect/train2/weights/best.pt")

results = model("/content/drive/MyDrive/Machine Learning Internship/Brain_Tumor_Detection_YOLO11_SAM2/Tumor_Detection/test/images/glioma_908_jpg.rf.cfc35d69a18fa2e2108e0b3bd341c50c.jpg", save = True)

sam = SAM("sam2_b.pt")


for result in results:
  class_ids = result.boxes.cls.tolist()
  if len(class_ids):
    boxes = result.boxes.xyxy
    sam_results = sam(result.orig_img, bboxes = boxes, verbose = False, save = True, device = 0)
    sam_results[0].show()
```
